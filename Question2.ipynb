```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Road.csv')
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Accident_severity')
plt.title('Distribution of Accident Severity')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Day_of_week', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Accidents by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Light_conditions')
plt.title('Accidents by Light Conditions')
plt.xlabel('Light Conditions')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age_band_of_driver', bins=10, kde=True)
plt.title('Age Distribution of Drivers')
plt.xlabel('Age Band of Driver')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Weather_conditions')
plt.title('Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
gender_counts = df['Sex_of_driver'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution of Drivers')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Vehicle_movement')
plt.title('Vehicle Movements During Accidents')
plt.xlabel('Vehicle Movement')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Cause_of_accident', order=df['Cause_of_accident'].value_counts().index)
plt.title('Most Common Accident Causes')
plt.xlabel('Accident Cause')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df)
plt.title('Age Distribution by Casualty Severity')
plt.xlabel('Casualty Severity')
plt.ylabel('Age Band of Driver')
plt.xticks(rotation=45)
plt.show()
```
